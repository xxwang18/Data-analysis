```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

filename = 'paraboloid_datadat.sec'
data = pd.read_csv(filename,sep='\t',header = None,names = ['x','y','f(x,y)'])
data
```

```python
data['x^2'] = data['x']**2
data['y^2'] = data['y']**2
data
```

```python
X = data[['x','y','x^2','y^2']]
Y = data['f(x,y)']
XT =X.T
```

```python
beta = XT.dot(Y).div(XT.dot(X))
beta
```
