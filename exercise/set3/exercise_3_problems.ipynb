# 1. Averaging spectral radio data [50 Points]

In this exercise you will work on a data set of artificial radio observation `radio-map.fits`. It is a map of 5x5 spectra with 201 measured frequency/velocity channels (channel width is unity). If you number all spectra from 1 to 25 we use the following scheme to assign spectra to positions on the map (spectrum 1 is the top left spectrum):

```python
from astropy.io import fits
import pandas as pd
import numpy as np

fits_image_filename = './radio-mapfits.sec'

radio_map = fits.open(fits_image_filename)
```

```python
radio_map.info()
```

```python
hdr = radio_map[0].header
hdr
```

```python
data = radio_map[0].data
```

**a.** Compute the total “integrated-intensity” map of the observations, i.e. integrate the spectra over the full spectral range for all positions and plot the 5x5 map in a suitable way. **10 Points**

```python
integrated_intensity_map = np.zeros(shape = (5,5))#create a shape(5,5) numpy array
for naxis2 in range(0,5):#integrate the spectra value for all postions 
    for naxis3 in range(0,5):
        for naxis1 in range(0,201):
            integrated_intensity_map[naxis2][naxis3] += data[naxis2][naxis3][naxis1]
print(integrated_intensity_map)
import matplotlib.pyplot as plt
plt.imshow(integrated_intensity_map)
plt.title('integrated-intensity map over the full spectral range')
plt.colorbar()
plt.show()#matrix plot of 'integrated-intensity' map
```

**b.** Compute two channel maps by integrating over the frequency channels 50-100 and 100-150. Compare the two maps. **10 Points**

```python
integrated_intensity_map_50_100 = np.zeros(shape = (5,5))
for naxis2 in range(0,5):
    for naxis3 in range(0,5):
        for naxis1 in range(50,100):
            integrated_intensity_map_50_100[naxis2][naxis3] += data[naxis2][naxis3][naxis1]
print(integrated_intensity_map_50_100)
plt.imshow(integrated_intensity_map_50_100)
plt.title('integrated-intensity map over the 50-100')
plt.colorbar()
plt.show()
```

```python
integrated_intensity_map_100_150 = np.zeros(shape = (5,5))
for naxis2 in range(0,5):
    for naxis3 in range(0,5):
        for naxis1 in range(100,150):
            integrated_intensity_map_100_150[naxis2][naxis3] += data[naxis2][naxis3][naxis1]
print(integrated_intensity_map_100_150)
plt.imshow(integrated_intensity_map_100_150)
plt.title('integrated-intensity map over 100-150')
plt.colorbar()
plt.show()
```

**c.** Compute the average spectrum, by averaging all 25 positions. **10 Points**

```python
average = np.zeros(shape=201)
for naxis1 in range(201):
    for naxis2 in range(0,5):
        for naxis3 in range(0,5):
            average[naxis1] += data[naxis2][naxis3][naxis1]
    average[naxis1]= average[naxis1]/25
plt.plot(average,label='average')
plt.title('average spectrum')
plt.xlabel('channel number')
plt.ylabel('intensity')
```

**d.** Plot every spectrum and overlay the average spectrum. Describe how the emission changes across the map. In particular how the line center position, the peak height and the line widths behave. **20 Points**

```python
for m in range(5):
    for n in range(5):
        plt.plot(data[m][n],label = 'spectrum ('+str(m)+','+str(n)+')')
        
plt.plot(average,'ro',label='average')
plt.title('every spectrum and overlay the average spectrum')
plt.xlabel('channels number')
plt.ylabel('intensity')
plt.legend(loc=(1.04,0.05), ncol=2)
```

# 2. Bayesian inference - conditional probabilities [50 Points]

Consider the proverbial bad penny, for which prior information has indicated that there is a probability of $P_r(B)=99$ that it is unbiased ("ok"); or a probability of $P$(B)=0.01 that it is double-headed ("dh").

**a.** Draw a tree diagram and label and label the nodes with the corresponding events and the vertices with the corresponding probabilities. **5 Points**

**b.** Suppose that the coin is fair, what is the probability P(A|B) of obtaining n heads in a row? **10 Points**

B is the event that the coin is fair, A is the event that obtaining n heads in a row. $$P(A|B) = (\frac {1}{2})^n$$

**c.** Now suppose, the coins is unfair, i.e. double-headed, what is the probability P(A|B) of obtaining n heads in a row in this case? **5 Points**

B is the event that the coin is unfair, A is the event that obtaining n heads in a row. $$ P(A|B)=1 $$

**d.** Give the probability P(A) = P(n heads in a row) given our initial priors. **10 Points**

  A is the event that obtaining n heads in a row, B is the event that the coin is fair. $$P(A)= P(A|B)\times P(B)+P(A|\overline{B}) \times P(\overline{B})={\frac {1}{2}}  ^n \times 0.99 + 0.01$$

**e.** Give the (Bayesian) posterior probability P(B|A) of obtaining n heads in a row. **10 Points**

$$P(B|A)= \frac {P(A|B) \times P(B)}{P(A)} = \frac { {\frac {1}{2}}^n \times 0.99}{{\frac {1}{2}}  ^n \times 0.99 + 0.01}$$

**f.** What are the odds on the penny being fair, i.e. what is the (Bayesian) posterior probability P(B|A) given this information, if we obtain 2 heads in a row? What is the probability if we obtain 7 heads in a row? In the last case (7 heads) how might we consider the fairness of the coin? Or of the experimenter who provided us with the prior information? **10 Points**

  Obtain 2 heads in a row: $$P(B|A)= \frac {P(A|B) \times P(B)}{P(A)} = \frac { {\frac {1}{2}}^2 \times 0.99}{{\frac {1}{2}}  ^2 \times 0.99 + 0.01} = 0.9612$$

  Obtian 7 heads in a row:$$P(B|A)= \frac {P(A|B) \times P(B)}{P(A)} = \frac { {\frac {1}{2}}^7 \times 0.99}{{\frac {1}{2}}  ^7 \times 0.99 + 0.01} = 0.4361$$
