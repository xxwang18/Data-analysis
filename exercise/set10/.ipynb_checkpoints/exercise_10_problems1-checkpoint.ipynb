# 1. Maximum Likelihood analysis [100 points]

## Flux density measurements

In this exercise we repeat an example from the lecture and  extend it: **Flux density measurements** (Lecture Note 11 p. 17).

Like in the lecture we suppose that we have flux density measurements at the frequencies,

  $$ f_i\ \in\ \big(0.4,\ 1.4,\ 2.7,\ 5,\ 10\big) \ \mathrm{GHz}, $$
  
with corresponding measured flux densities,

  $$ 1.855,\ 0.640,\ 0.444,\ 0.22,\ 0.102\ \mathrm{flux\ units}. $$

Let the frequencies be $f_i$ and the data $S_i$. The measurements follow a power-law slope of $-1$, on top of which a $10\%$ Gaussian noise level is added. The noise level is denoted $\epsilon$ and the model for the flux density as a function of frequency is $kf^{-\gamma}$. Assuming we know the noise level and distribution, each term in the likelihood product is of the form:

$$ 
\frac{1}{\sqrt{2\pi}\epsilon{}k{}f_i^{-\gamma}} \mathrm{exp} \Big( - \frac{(S_i-k{}f_i^{-\gamma})^2)}{2(\epsilon{}k{}f_i^{-\gamma})^2} 
\Big)
$$

We are therefore fitting Gaussians to the errors.

**a)** Plot the measured spectrum on a log-log scale. **10 points**

```python
frequency = np.array([0.4, 1.4, 2.7, 5, 10])
flux_densities = np.array([1.855, 0.640, 0.444, 0.22, 0.102])
```

```python
fig, ax = plt.subplots()
ax.plot(frequency, flux_densities, linewidth=0, marker='s', label='Data points')
ax.set_xlabel('Frequency(GHz)')
ax.set_ylabel('Flux Density (flux units)')
ax.legend(facecolor='white')
ax.set_xscale('log')
ax.set_yscale('log')
```

**b)** Compute the logarithmic likelihood for a range of $k$ and $\gamma$ values in the range of the maximum likelihood. Normalize the likelihood. Produce a 2-D plot of the logarithmic likelihood for each calculated pair of $k$ and $\gamma$. **20 points**

```python
def likelihood_modelA(epsilon, k, f_i, S_i, gamma):
    term = (1/(np.sqrt(2*np.pi)*epsilon*k*np.power(f_i, -1*gamma))) * np.exp(-((S_i - k*np.power(f_i, -1*gamma))**2)/(2*(epsilon*k*np.power(f_i, -1*gamma))**2))
    return term

def log_LL(observations, epsilon, k_range, gamma_range, frequencies):
    f_i = frequencies
    k,gamma = np.meshgrid(k_range,gamma_range)
    L = np.ones(shape=[len(k), len(gamma)])
    log_L = np.zeros(shape=[len(k), len(gamma)])
    for index, S_i in enumerate(observations):
        L *= likelihood_modelA(epsilon, k, f_i[index], S_i, gamma)
    L = L/np.sum(L)
    #L = L/np.trapz(L, k_range) #NORMALIZATION
    #L = L/np.trapz(L, gamma_range)
    #print(np.sum(L))
    log_L = np.log(L)
    return L, log_L
        
```

```python
epsilon = 0.2
k_range = np.linspace(0.45,2.8,100)
gamma_range = np.linspace(0.2,2.0,100)
likelihood, log_likelihood = log_LL(flux_densities, epsilon, k_range, gamma_range, frequency)
#print(log_likelihood)

#PLOT
k,gamma = np.meshgrid(k_range,gamma_range)
plt.contour(k,gamma,log_likelihood, levels = [-25,-20,-15,-10,-5,0,5])
plt.colorbar()
plt.title('Logarithmic Likelihood Contours')
plt.xlabel('k')
plt.ylabel(r'$\gamma$')
```

**c)** Now look at the probability of one parameter independent of the other parameter. This is called marginalization. In other words we have the posterior probability $\mathcal{P}_r(k,\gamma|S_i)$ and can form:

$$\mathcal{P}_r(k|S_i) = \int{\mathcal{P}_r(k,\gamma|S_i) d\gamma}$$

Compute the marginal probability distributions for $k$ and $\gamma$ and plot them individually. **20 points**

The likelihood function is the joint prob

```python
P_k_given_S = np.trapz(likelihood, gamma_range)
P_gamma_given_S = np.trapz(likelihood, k_range)
P_k_given_S = P_k_given_S/np.trapz(P_k_given_S, k_range)   #NORMALIZATION
P_gamma_given_S = P_gamma_given_S/np.trapz(P_gamma_given_S, gamma_range )   #NORMALIZATION

fig,ax = plt.subplots(1,2,figsize=(12, 4))
ax[0].plot(k_range, P_k_given_S)
ax[0].set_title('The marginal distribution of k')
ax[0].set_xlabel('k')
ax[0].set_ylabel(r'$P_r(k|S_i)$')
ax[1].plot(gamma_range,P_gamma_given_S)
ax[1].set_title(r'The marginal distribution of $\gamma$')
ax[1].set_xlabel('$\gamma$')
ax[1].set_ylabel(r'$P_r(\gamma|S_i)$')
```

**d)** Add an *artificial* 0.4 flux units to the flux measurements, then use it to calculate two models: 

 - Model A is the same as above with no offsets. 

 - Model B uses a model for the flux densities of the form of $\beta+kf^{-\gamma}$. Thus each likelihood term is:

$$
\frac{1}{\sqrt{2\pi}\epsilon{}k{}f_i^{-\gamma}} \mathrm{exp} \Big( - \frac{(S_i-(\beta + k{}f_i^{-\gamma}))^2}{2(\epsilon{}k{}f_i^{-\gamma})^2} 
\Big)
$$

Plot the contours of the computed logarithmic likelihoods for each Model into the same plot. Compare both models with each other. Which one gives better results? Explain why. **30 points**

```python
flux_densities += 0.4
def likelihood_modelB(epsilon, k, f_i, S_i, gamma, beta):
    term = (1/(np.sqrt(2*np.pi)*epsilon*k*np.power(f_i, -1*gamma))) * np.exp(-( (S_i - (beta + k*np.power(f_i, -1*gamma)))**2 )/(2*(epsilon*k*np.power(f_i, -1*gamma))**2))
    return term

def log_LL_modelB(observations, epsilon, k_range, gamma_range, frequencies, beta_range):
    f_i = frequencies
    beta, k, gamma = np.meshgrid(beta_range, k_range, gamma_range)
    L = np.ones(shape=[len(beta), len(k), len(gamma)])
    log_L = np.zeros(shape=[len(beta), len(k), len(gamma)])
    for index, S_i in enumerate(observations):
        L *= likelihood_modelB(epsilon, k, f_i[index], S_i, gamma, beta)
    L = L/np.sum(L)
    L_integrated_over_beta = np.trapz(L, beta_range)
    #print(np.sum(L_integrated_over_beta ))
    L_integrated_over_beta = L_integrated_over_beta/np.sum(L_integrated_over_beta)
    log_L = np.log(L_integrated_over_beta)
    return L, log_L
```

```python
epsilon = 0.2
beta_range = np.linspace(0.1, 1, 50)
k_range = np.linspace(0.1,2.0,50)
gamma_range = np.linspace(-0.1,1.5,50)

likelihood_A, log_likelihood_A = log_LL(flux_densities, epsilon, k_range, gamma_range, frequency)
#print(log_likelihood_A)
likelihood_B, log_likelihood_B = log_LL_modelB(flux_densities, epsilon, k_range, gamma_range, frequency, beta_range)
#print(log_likelihood_B)

#PLOT

fig, ax = plt.subplots(figsize=(6, 6))
modelA = ax.contour(k_range,gamma_range,log_likelihood_A,linestyles = 'solid',colors = ['black'])#label = r'Model A: $P_r(K,\gamma)|S_i$'
ax.clabel(modelA,inline = 1,fontsize = 10)
modelB = ax.contour(k_range,gamma_range,log_likelihood_B,linestyles = 'dashed',colors = ['red'] )#label = r'Model B: $\int P_r(\beta,k,\gamma|S_i)d\beta$'
ax.clabel(modelB,inline = 1,fontsize = 10)
ax.set_title('Logarithmic Likelihoods')
ax.set_xlabel('k')
ax.set_ylabel(r'$\gamma$')
ax.text(x=0.2, y=1.37,s=r'Model A: $P_r(k,\gamma)|S_i$',fontsize=12)
ax.text(x=0.2, y=1.3,s=r'Model B: $\int P_r(\beta,k,\gamma|S_i)d\beta$',fontsize=12,color='red')
```

**e)** Compute the marginal probability distributions for Model B for the three parameters: $\beta$, $\gamma$, and $k$ and plot them. **20 points**

```python
fig,ax = plt.subplots(1,3,figsize=(16, 4))
p_k = np.trapz(np.trapz(likelihood_B, beta_range),gamma_range)
p_gamma = np.trapz(np.trapz(likelihood_B, beta_range),k_range)
p_beta = np.trapz(np.trapz(likelihood_B,gamma_range),k_range)
p_k = p_k/np.sum(p_k)
p_gamma = p_gamma/np.sum(p_gamma)
p_beta = p_beta/np.sum(p_beta)
ax[0].plot(k_range,p_k)
ax[0].set_title('the marginal distributions of k')
ax[0].set_xlabel('k')
ax[0].set_ylabel(r'$P_r(k|S_i)$')
ax[1].plot(gamma_range,p_gamma)
ax[1].set_title(r'the marginal distributions of $\gamma$')
ax[1].set_xlabel('$\gamma$')
ax[1].set_ylabel(r'$P_r(\gamma|S_i)$')
ax[2].plot(beta_range,p_beta)
ax[2].set_title(r'the marginal distributions of $\beta$')
ax[2].set_xlabel(r'$\beta$')
ax[2].set_ylabel(r'$P_r(\beta|S_i)$')
```
